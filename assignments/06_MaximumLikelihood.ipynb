# Обобщенная линейная модель. Метод максимального правдоподобия

Правдоподобие: вероятность наблюдать наши ответы с теми примерами которые есть при заданном параметре W. И наша задача найти те W при которых это правдоподобие максимально, то есть чтобы вероятность получить такие данные была как можно выше.
$$L(W) = p(y|X;W)$$ 
$L$ - likelihood. 
Откуда брать эту $p$? Это вероятность получить весь набор в совокупности, в таком виде задача очень сложна. НО!  
Теперь супер трюк: считаем что каждый пример сам по себе и не зависит от остальных, тогда мы можем представить вероятность как произведение

$$L(W) = \prod p(y_i|x_i;W)$$
теперь нам достачно знать генерирующее распределение такого случайного процесса, если для него есть формула и оно хорошее в каком-то смысле, то можем просто подставить в формулу параметры и решить уравнение. Уровень современной математики таков, что уравнения решать научились супер сложные, главное записать их.



## Семейство экспоненциальных распределений

Есть широкий класс очень хороших экспоненциальных распределений. Он хороший, потому что для него хорошо считается функция правдоподобия и можно решить огромное количество задач. Именно оно и используется в так называемых обобщенных линейных моделях. Рассмотрим его в общем виде и самые популярные


$$p(y;\eta) = b(y) \exp (\eta^T T(y) - a(\eta))$$

Параметры:
  - $y$ - наши данные
  - $\eta$ - параметр(ы) распределения
  - T(y)  - достаточная статистика для у (в нашем случае это будет просто единичное отображение T(y) = y)
   -a(\eta) = log-partition function ( нас не сильно интересует)
   
   
### Обобщенная линейная модель (GLM):
 - Выход моделируется распределением из экспоненциального семейства распределений
 - $\eta = W^TX$ - вот она линейная функция относительно параметров W
 - $W$ -  параметры, максимизирующие правдоподобие
 - $h_W(x) = E(y;W^TX) = g(\eta)$   - математическое ожидание при найденном наборе параметров, $g(\eta)$ - функция отклика,
 а обратная к ней называется функцией связи (link function) и ответа.
 
 То есть в итоге вы нашли параметры W,  получили какой-то новый вход, посчитали скалярное произведение $W^TX$, а чтобы получить ответ еще применили функцию g.  Сейчас посмотрим на примерах
 
 

##  Гауссовское распределение (оно же нормальное): классическая линейная регрессия
Рассмотрим случай  с фиксированной дисперсией (всегда можем добиться нормировкой)
 $$p(y;\mu)=\frac{1}{\sqrt{2\pi}}\frac{\exp-(y_i - \mu)^2}{2}$$
 
 $$\exp-(y -\mu)^2 = \exp(-y^2/2+\mu y -1/2\mu^2)=\exp(-y^2/2)\exp(\mu y - 1/2\mu^2) $$



$$p(y;\eta) = b(y) \exp (\eta^T T(y) - a(\eta))$$
$$p(y;\mu) =\exp(-y^2/2)\exp(\mu y - 1/2\mu^2)  $$

То есть здесь $T(y) = y$, $\eta = \mu = W^TX$

## Биномиальное распределение (Логистическая регрессия)

$$p(y;\theta) = \theta^y(1-\theta)^{1-y}$$

Как привести к экспоненциальному виду?

$$p(y;\theta) = \exp(\log\frac{\theta}{1-\theta}y+\log(1-\theta))$$

$$\theta = \frac{1}{1+\exp(-W^TX)}$$ 

$$E(y;W^TX) = g(W^TX)$$
это логистическая функция, еще известная как сигмоида, она умеет превращать естественную ось в диапазон
